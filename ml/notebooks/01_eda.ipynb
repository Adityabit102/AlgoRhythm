# AlgoRhythm — EDA

The twelve story beats: what does the data say makes a hit?
Run on `data/processed/tracks.csv` (synthetic until real data lands).


```python
import sys; sys.path.append('../src')
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
df = pd.read_csv('../data/processed/tracks.csv')
df.shape, df.is_hit.mean()
```

## 01 · The Danceability Threshold


```python
sns.kdeplot(data=df, x='danceability', hue='is_hit', common_norm=False)
plt.title('Danceability: hits vs non-hits'); plt.show()
```

## 02 · The Friday Effect


```python
df['dow'] = pd.to_datetime(df.release_date).dt.dayofweek
df.groupby(df.dow==4).is_hit.mean()
```

## 06 · Tempo Distribution of Hits


```python
sns.histplot(data=df[df.is_hit==1], x='tempo', bins=40); plt.show()
```

## Correlation of features with the hit label


```python
num = df.select_dtypes('number')
num.corr()['is_hit'].sort_values(ascending=False)
```
